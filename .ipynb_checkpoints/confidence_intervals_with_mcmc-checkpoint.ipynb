# **Define system of differential equaitions**

```python
def virus_attack_1(y,t, k_1,k_8, ro_8, ro_V, r):
#     oprimized parameters
#     k_1,k_8, ro_8, ro_V = p  
#     states
    y1,y2, y3, y4 = y 
#     parameters
    s_3=5
#     ro_8=3.3*0.0001
    mu_C=0.015
    s_1=5
#     k_1=3.87*(0.001) 
    mu_T=0.02
    S=55.6
    p_1=0.16
#     r=0.02
    maxi=1500
    mu_ti=0.28    
#     k_8=4.5*0.0001
    ro_4=0.002
#     ro_V=100
    mu_V=3.0  
    k_10=7.79*1e-6
    dy1dt = s_1-k_1*y1*y4-mu_T*y1+r*y1*(1-(y1+y2)/maxi)
    dy2dt = k_1*y1*y4+ro_4*y2*(1-(y1+y2)/maxi)-k_8*y3*y2-mu_ti*y2
    dy3dt=s_3+ro_8*y2*y3-mu_C*y3
    dy4dt=ro_V*y2-k_10*y1*y4-mu_V*y4
    return [dy1dt,dy2dt, dy3dt, dy4dt]
```

## **Add objective function**

```python
def sim_model(xdata, x):
    k_1,k_8, ro_8, ro_V, r =x
    CD_p=[]
    CD_ip=[]
    CTL_p=[]
    V_p=[]
    t=xdata.reshape(n)
    for i in range(0,n-1):
        a=t[i]
        b=t[i+1]
        ts=np.linspace(a,b,b-a+1)
#         ts=[t[i], t[i+1]]
        inputs = (k_1,k_8, ro_8, ro_V , r)
        y = odeint(virus_attack_1,[CDp[i],Tip[i], CCp[i], Vp[i]],ts,args=inputs)     
        CDp[i+1] = y[-1,0]
        Tip[i+1]=y[-1,1]
        CCp[i+1]=y[-1,2]
        Vp[i+1] = y[-1,3]
        CD_p+=list(y[:,0][:-1])
        CD_ip+=list(y[:,1][:-1])
        CTL_p+=list(y[:,2][:-1])
        V_p+=list(y[:, 3][:-1])
    CD_p+=[y[:,0][-1]]
    CD_ip+=[y[:,1][-1]]
    CTL_p+=[y[:,2][-1]]
    V_p+=[y[:, 3][-1]]
    return np.log(CDp), np.log(Vp)   

def objective(x, data):
    xdata=data.xdata[0]

    ss=np.zeros([len(data.xdata)])
    CD=data.ydata[0].reshape(n)
    V=data.ydata[1].reshape(n)
    CDp, Vp  = sim_model(xdata, x)
    ss[0]=sum(np.power(CDp-CD, 2))
    ss[1]=sum(np.power(Vp-V,2))
    return ss
```

```python
# set patient
patient=3

virus['V']=np.power(10,virus[str('VP')+str(patient)])/2000
n = virus[str('TP')+str(patient)].shape[0]# time points to plot
tf =virus[str('TP')+str(patient)][n-1]  # final time
# Use initial condition
y0d = [virus[str('CDP')+str(patient)].values[0],0, 617, virus['V'].values[0]]

CDp = np.ones(n)*y0d[0]
Tip=np.ones(n)*y0d[1]
CCp=np.ones(n)*y0d[2]
Vp = np.ones(n)*y0d[3]
t=virus[str('TP')+str(patient)].values-virus[str('TP')+str(patient)].values[0]

```

## **Define data**

```python
x1=virus[str('TP')+str(patient)].values-virus[str('TP')+str(patient)].values[0]
y1=np.log(virus[str('CDP')+str(patient)].values)
x2=virus[str('TP')+str(patient)].values-virus[str('TP')+str(patient)].values[0]
y2=np.log(virus['V'].values)
mcstat = MCMC()
mcstat.data.add_data_set(x1, y1)
mcstat.data.add_data_set(x2, y2)
mcstat.simulation_options.define_simulation_options(
    nsimu=int(6.0e4),
    updatesigma=True,
    method='dram')
mcstat.model_settings.define_model_settings(sos_function=objective)
mcstat.parameters.add_model_parameter(name='k_1', theta0=5.64e-06, minimum=1e-6, maximum=1e-5)
mcstat.parameters.add_model_parameter(name='k_8', theta0=0.0021, minimum=0.0008, maximum=0.007)
mcstat.parameters.add_model_parameter(name='ro_8', theta0=0.0069, minimum=0.004, maximum=0.01)
mcstat.parameters.add_model_parameter(name='ro_V', theta0=2356.53, minimum=2000, maximum=3000)
mcstat.parameters.add_model_parameter(name='r', theta0=0.0223, minimum=0.001, maximum=0.04)
```

## **Run MCMC simulations**

```python
mcstat.run_simulation()
results = mcstat.simulation_results.results
chain = results['chain']
s2chain = results['s2chain']
names = results['names']
# define burnin
burnin = int(results['nsimu']/2)
# display chain statistics
mcstat.chainstats(chain[burnin:, :], results)
```

## **Generate confidence intervals**

```python
# generate prediction intervals
from pymcmcstat import propagation as up
from pymcmcstat.MCMC import DataStructure

def predmodel_1(q, data):
    return sim_model1(data.xdata[0], q)
def predmodel_2(q, data):
    return sim_model2(data.xdata[0], q)

pdata = []
intervals = []
nds = len(mcstat.data.xdata)

for ii in range(nds):
    pdata.append(DataStructure())
    print(pdata)
    pdata[ii].add_data_set(mcstat.data.xdata[ii], mcstat.data.ydata[ii])
    if ii==0:
      intervals.append(up.calculate_intervals(chain, results, pdata[ii], predmodel_1,
                                  waitbar=True, s2chain=s2chain[:, ii]))
    else:
      intervals.append(up.calculate_intervals(chain, results, pdata[ii], predmodel_2,
                                  waitbar=True, s2chain=s2chain[:, ii]))
```
